```python
# 붓꽃 데이터 셋
iris = sns.load_dataset("iris")
print(iris)
```

```python
# 타이타닉 데이터 셋
titanic = sns.load_dataset("titanic")
titanic
```

```python
# 팁 데이터 셋
tips = sns.load_dataset("tips")
tips
```

```python
# fmri 데이터 셋
fmri = sns.load_dataset("fmri")
fmri
```

```python
#scatterplot : 산점도 scatter(x, y, data)
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
#hue : 의미에 따라 점의 색을 변경
#style: 모양 구분
sns.scatterplot(x='total_bill', y= 'tip', data = tips, hue = 'day', style = 'time')
```

```python
#lineplot : 데이터가 연속적일 경우 주로 사용 .lineplot(x,y,data)
sns.lineplot(x='timepoint', y='signal', data=fmri)
```

```python
# 위의 그래프에서 색이 칠해져 있는 부분은 신뢰구간(confidene interval)로 ci 파라미터로 조절 가능
# hue와 style 옵션을 사용할 수 있음

sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event',ci=None)
```

```python
# relplot .relplot(x,y,kind,data)
# kind 파라미터에 scatter나 line으로 형식 선택 가능 (default=scatter)
# scatterplot과 lineplot은 AxeSubplot을 반환하지만, relplot은 FaceGrid를 반환
# FaceGrid를 반환하는 경우 여러 그래프를 한 번에 그릴수 있음
# hue와 style 옵션 모두 사용 가능

sns.relplot(x='total_bill',y='tip',kind='scatter',hue ='time',data=tips)
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',hue ='event', style ='event',ci = None,data=fmri)
```

```python
#distribution : 변수 하나 혹은 두개의 값 분포를 나태내기 위한 그래프
#histplot : 히스토그램 그래프를 그리는 함수로, 데이터의 빈도를 나타낼 때 사용 #histplot(x,data)

sns.histplot(x='sepal_length',data=iris)
```

```python
#hue 옵션으로 레이블 구분
#multiple 옵션으로 hue로 분류된 레이블을 중첩되게 설정 가능

sns.histplot(x='sepal_length',data=iris,hue='species')
```

```python
#multiple 옵션으로 hue로 분류된 레이블을 중첩되게 설정 가능
sns.histplot(data=iris, x='sepal_length', hue='species', multiple='stack')
```

```python
#hisplot 2차원 배열로 나타내기 -> count정보는 안나옴 ->multiple줘도 안바뀜
sns.histplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
```

```python
# kdeplot
# 커널 밀도 추정(kernal density estimation) 그래프로,
# 히스토그램이 절대량(count)을 표현한다면 kdeplotdms 상대량(비율)을 시각화
# 히스토그램과 마찬가지로 한개 혹은 두개의 변수에 대한 분포를 그릴 수 있음

sns.kdeplot(x = 'sepal_length',data=iris)
```

```python
sns.kdeplot(x ='sepal_length',data=iris,hue='species')
```

```python
sns.kdeplot(x = 'sepal_length',y='sepal_width',data=iris,hue='species')
```

```python
# histplot과 kedplot을 함께 사용하는 경우
sns.histplot(x='sepal_length',data=iris,hue='species',multiple='stack',kde=True)
```

```python
# rugplot
# 데이터 값을 x축과 y축 위에 그려주는 그래프 -> 단독으로 사용되기보다는 다른 그래프와 함께 사용된다(i.e. kdeplot)
# 다른 그래프를 보완해주는 느낌

sns.rugplot(x='sepal_length',y='sepal_width',data=iris,hue='species')
```

```python
sns.kdeplot(x='sepal_length',y='sepal_width',data=iris,hue='species')
sns.rugplot(x='sepal_length',y='sepal_width',data=iris,hue='species')
```

```python
# ecdfplot
# 분포를 누적화해 보여줌

sns.ecdfplot(x = 'sepal_length',data = iris)
```

```python
# displot
# 앞선 함수들의 통합 개념들로, relplot과 유사함
# row와 col 파라미터로 subplot 생성 가능
# rug 옵션으로 rugplot의 표시 여부를 선택 가능

sns.displot(data=iris, x='sepal_length', hue='species') # kind default = hist
```

```python
#col은 행으로 보여줌
sns.displot(x='sepal_length',data=iris,hue='species',col ='species')
```

```python
#row는 열단위로 보여줌
sns.displot(data=iris,x='sepal_length',kind='kde',hue='species',row='species',rug=True)
```
